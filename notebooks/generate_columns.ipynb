Split into longitude and latitude for future processing.

Clean up invalid locations from parse data. We just clean them as we won't lose a lot of records and these records might not be that important for us.

````python
parsed_data_cleaned = parsed_data.dropna(subset=['location'])
parsed_data_cleaned.shape
````

````python
# Function to split location into latitude and longitude
def split_location(location: str):
    lat, lon = location.split(',')
    return float(lat), float(lon)

# If parsed_data_cleaned is a slice, make a copy to avoid SettingWithCopyWarning
parsed_data_cleaned = parsed_data_cleaned.copy()

# Apply the function to split the 'location' column into 'latitude' and 'longitude' columns
parsed_data_cleaned[['latitude', 'longitude']] = parsed_data_cleaned['location'].apply(
    lambda x: pd.Series(split_location(x))
)
````

````python
parsed_data['location'].dtype
````

````python
invalid_longitudes = parsed_data_cleaned.where(~((parsed_data_cleaned['longitude'] >= -180) & (parsed_data_cleaned['longitude'] <= 180)))
invalid_longitudes.dropna()
````

````python
invalid_latitudes = parsed_data_cleaned.where(~((parsed_data_cleaned['latitude'] >= -90) & (parsed_data_cleaned['latitude'] <= 90)))
invalid_latitudes.dropna()
````

````python
location_coordinates = [(lat, long) for lat, long in zip(parsed_data_cleaned['latitude'].to_list(), parsed_data_cleaned['longitude'].to_list())]
location_coordinates[:10]
````

````python
real_location_info = rg.search(location_coordinates, mode=1) # default mode = 2 # we use mode=1 to avoid inconsistencies caused by multi threading
real_location_info
````

````python
# Use .loc to modify the DataFrame to avoid SettingWithCopyWarning
parsed_data_cleaned.loc[:, 'actual_city_name'] = [result['name'] for result in real_location_info]
parsed_data_cleaned.loc[:, 'actual_admin1'] = [result['admin1'] for result in real_location_info]
parsed_data_cleaned.loc[:, 'actual_admin2'] = [result['admin2'] for result in real_location_info]
parsed_data_cleaned.loc[:, 'actual_cc'] = [result['cc'] for result in real_location_info]
````

````python
# Fill NaN/None values in 'english_common_name' with 'unidentified'
parsed_data_cleaned['english_common_name'] = parsed_data_cleaned['english_common_name'].fillna('unidentified')

# Combine the specified columns into the 'roboflow_file_name' column
parsed_data_cleaned['roboflow_file_name'] = (
    parsed_data_cleaned['observation_id'].astype(str) + "_" +
    parsed_data_cleaned.index.astype(str) + "_" +
    parsed_data_cleaned['english_common_name'].str.lower().str.replace('/', '').str.replace(' ', '_') + "_" +
    parsed_data_cleaned['taxon_name'].str.lower().str.replace('/', '').str.replace(' ', '_') + "_" +
    parsed_data_cleaned['observed_on_year'].astype(str) + "_" +
    parsed_data_cleaned['observed_on_month'].astype(str) + "_" +
    parsed_data_cleaned['observed_on_day'].astype(str) + "_" +
    parsed_data_cleaned['observed_on_hour'].astype(str) + "_" +
    parsed_data_cleaned['observed_time_zone'].str.lower().str.replace('/', '_').str.replace(' ', '_') + "_" +
    parsed_data_cleaned['actual_city_name'].str.lower().str.replace(' ', '_') + "_" +
    parsed_data_cleaned['actual_admin1'].str.lower().str.replace(' ', '_') + "_" +
    parsed_data_cleaned['actual_admin2'].str.lower().str.replace(' ', '_') + "_" +
    parsed_data_cleaned['actual_cc']
)

# View the resulting DataFrame
parsed_data_cleaned[['roboflow_file_name']].head()
````
